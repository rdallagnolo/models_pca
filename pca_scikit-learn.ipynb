```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_model = pca.fit_transform(scaled_X)
```

```python
pca_model
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_model[:,0],pca_model[:,1],c=cancer_dictionary['target'])
```

```python
pca.components_
```

```python
df_components = pd.DataFrame(pca.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_components
```

```python
plt.figure(figsize=(20,3))
sns.heatmap(df_components,annot=True)
```

```python
pca.explained_variance_ratio_
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    
    pca=PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_)) 
    
    
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel('n of components')
plt.ylabel('variance explained')
```
